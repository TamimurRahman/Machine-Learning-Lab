```python
string = input('Enter a string: ')
string_list = list(string)
frequency = {}
for i in string_list:
    if i in frequency:
      frequency[i]+=1
    else:
      frequency[i]=1
for key,value in frequency.items():
   print(f'{key} : {value}',end=" ")

print(string_list)
```
